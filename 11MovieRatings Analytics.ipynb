```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style("dark",{"axes.facecolor":"black"})
f,axes=plt.subplots(2,2,figsize=(15,15)) 

#plot[0,0]
k2=sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,shade=True,shade_lowest=True,cmap='YlOrRd_r',ax=axes[0,0] )

#plot[0,1]
k3=sns.kdeplot(movies.BudgetMillions,movies.CriticRating,shade=True,shade_lowest=True,cmap='YlOrRd_r', ax=axes[0,1]) 

k2.set(xlim=(-20,160))
k3.set(xlim=(-20,160))

#plot[1,0]

z=sns.violinplot(data=movies,x='Year',y='BudgetMillions',ax=axes[1,0],palette='YlOrRd')

#plot[1,1]

klb= sns.kdeplot(movies.CriticRating , movies.AudienceRating,shade=True,shade_lowest=False,cmap='inferno',ax=axes[1,1])
```

```python
#Thematic edits
```

```python
list1 = []
mylabels=[]
sns.set_style("whitegrid")
fig,ax=plt.subplots() # so as to set size easily
fig.set_size_inches(11.7,8.27) #size of a4 paper
for i in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == i].BudgetMillions)
    mylabels.append(i)

plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels,color=["orange","cyan","yellow","purple","blue","green","violet"])
plt.title("Movie Budget Distribution",fontsize=35,color='DarkBlue',fontname='Serif')
plt.ylabel("Number of Movies",color='DarkBlue',fontsize=25)
plt.xlabel("Budget",color='Green',fontsize=25)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(prop={'size':20},frameon=True,fancybox=True,shadow=True,framealpha=1) #framealpha :transparency
plt.show()
```
